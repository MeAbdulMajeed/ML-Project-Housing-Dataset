```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.mainroad = le.fit_transform(dfle.mainroad)
dfle.guestroom = le.fit_transform(dfle.guestroom)
dfle.basement = le.fit_transform(dfle.basement)
dfle.hotwaterheating = le.fit_transform(dfle.hotwaterheating)
dfle.airconditioning = le.fit_transform(dfle.airconditioning)
dfle.prefarea = le.fit_transform(dfle.prefarea)
dfle.furnishingstatus = le.fit_transform(dfle.furnishingstatus.map({'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0}))
dfle
```

```python
dfle.tail(15)
```

```python
dfle.info()
```

```python
X = dfle[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
y = dfle.price
```

```python
dfle.ndim
```

```python
X.ndim
```

```python
y.ndim
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Train the **Model**

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

## **Predict the test Results**

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2_score(y_test, y_pred)
```
